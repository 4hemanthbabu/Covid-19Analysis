# Total Cases in Major 7 Countries

```python
China_details = File.loc[File['location'] == 'China']
print(China_details)
US_details = File.loc[File['location'] == 'United States']
print(US_details)
UK_details = File.loc[File['location'] == 'United Kingdom']
print(UK_details)
AUS_details = File.loc[File['location'] == 'Australia']
print('AUS_details')
Brazil_details = File.loc[File['location'] == 'Brazil']
print('Brazil_details')
RUS_details = File.loc[File['location'] == 'Russia']
print(RUS_details)
```

```python

style.use('ggplot')
India_TC = India_details.groupby('date') 
India_Total_Cases = India_TC.sum()
China_TC = China_details.groupby('date')
China_Total_Cases = China_TC.sum()
US_TC = US_details.groupby('date')
US_Total_Cases = US_TC.sum()
Brazil_TC = Brazil_details.groupby('date')
Brazil_Total_Cases = Brazil_TC.sum()
UK_TC = UK_details.groupby('date')
UK_Total_Cases = UK_TC.sum()
RUS_TC = RUS_details.groupby('date')
RUS_Total_Cases = RUS_TC.sum()
AUS_TC = AUS_details.groupby('date')
AUS_Total_Cases = AUS_TC.sum()

y = India_Total_Cases['total_cases'].to_numpy()
x = nu.arange(len(y))
y = nu.array(y)
y1 = China_Total_Cases['total_cases'].to_numpy()
x1 = nu.arange(len(y1))
y1 = nu.array(y1)
y2 = US_Total_Cases['total_cases'].to_numpy()
x2 = nu.arange(len(y2))
y2 = nu.array(y2)
y3 = Brazil_Total_Cases['total_cases'].to_numpy()
x3 = nu.arange(len(y3))
y3 = nu.array(y3)
y4 = UK_Total_Cases['total_cases'].to_numpy()
x4 = nu.arange(len(y4))
y4 = nu.array(y4)
y5 = AUS_Total_Cases['total_cases'].to_numpy()
x5 = nu.arange(len(y5))
y5 = nu.array(y5)
y6 = RUS_Total_Cases['total_cases'].to_numpy()
x6 = nu.arange(len(y5))
y6 = nu.array(y6)



lw = 3
plt.plot(x,y, '-.', label = 'India', color = 'r', linewidth = lw)
plt.plot(x1,y2, '-.', label = 'China', color = 'g', linewidth = lw)
plt.plot(x2,y2, '-.', label = 'Russia', color = 'b', linewidth = lw)
plt.plot(x3,y3, '-.', label = 'United States', color = 'c', linewidth = lw)
plt.plot(x4,y4, '-.', label = 'United Kingdom', color = 'm',linewidth = lw)
plt.plot(x5,y5, '-.', label = 'Brazil', color = 'k', linewidth = lw)

plt.legend()
plt.suptitle('Total Cases in Major 7 Countries')
plt.xlabel('Days')
plt.ylabel('Total_Cases')

plt.show()
```
